```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
from sklearn.preprocessing import StandardScaler

# Assuming `data_` is already defined as your dataset
# Assuming `model_dict` is the mapping of constructs to columns
model_dict = {
    'GA': ['GA1', 'GA2', 'GA3', 'GA4', 'GA5'],
    'GL': ['GL1', 'GL2', 'GL3', 'GL4', 'GL5'],
    'GP': ['GP1', 'GP2', 'GP3', 'GP4'],
    'AT': ['AT1', 'AT2'],
    'PV': ['PV1', 'PV2', 'PV3'],
    'AW': ['AW1', 'AW2'],
    'PI': ['PI1', 'PI2', 'PI3']
}

# Output dictionary to store results for each construct
convergent_analysis_results = {}

# Loop through each construct and its associated columns
for construct, columns in model_dict.items():
    # Extract data for the construct
    construct_data = data_[columns]
    
    # Step 1: Standardize the data for VIF calculation
    scaler = StandardScaler()
    construct_data_scaled = scaler.fit_transform(construct_data)
    
    # Step 2: Compute Variance Inflation Factor (VIF)
    vif = pd.DataFrame({
        "Variable": columns,
        "VIF": [np.linalg.inv(np.corrcoef(construct_data_scaled.T))[i, i] for i in range(construct_data_scaled.shape[1])]
    })
    
    # Step 3: Conduct Exploratory Factor Analysis (EFA)
    fa = FactorAnalyzer(n_factors=1, rotation=None)  # 1-factor solution
    fa.fit(construct_data)
    
    # Extract factor loadings
    loadings = fa.loadings_
    
    # Step 4: Compute Average Variance Extracted (AVE)
    squared_loadings = loadings**2
    AVE = np.mean(squared_loadings)
    
    # Step 5: Compute Composite Reliability (CR)
    loadings_sum = np.sum(loadings)
    CR = (loadings_sum**2) / ((loadings_sum**2) + np.sum(1 - squared_loadings))
    
    # Step 6: Compute Cronbach's Alpha (CA)
    k = construct_data.shape[1]  # Number of items
    item_variances = construct_data.var(axis=0, ddof=1)
    total_variance = construct_data.sum(axis=1).var(ddof=1)
    CA = (k / (k - 1)) * (1 - (item_variances.sum() / total_variance))
    
    # Step 7: Compute Within Correlation (HTMT Methodology)
    correlation_matrix = construct_data.corr()  # Compute pairwise correlations
    upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))  # Extract upper triangle
    within_correlation = upper_triangle.stack().mean()  # Average of non-NaN values
    
    # Prepare results
    factor_loadings = pd.DataFrame(loadings, index=columns, columns=["Factor1"])
    results = {
        "Factor Loadings (FL)": factor_loadings,
        "Average Variance Extracted (AVE)": AVE,
        "Composite Reliability (CR)": CR,
        "Cronbach Alpha (CA)": CA,
        "Variance Inflation Factor (VIF)": vif,
        "Within Correlation": within_correlation  # Add the within correlation
    }
    
    # Store results for this construct
    convergent_analysis_results[construct] = results

# Output the results for each construct
for construct, result in convergent_analysis_results.items():
    print(f"Results for {construct}:")
    print(f"Factor Loadings (FL):\n{result['Factor Loadings (FL)']}")
    print(f"Average Variance Extracted (AVE): {result['Average Variance Extracted (AVE)']}")
    print(f"Composite Reliability (CR): {result['Composite Reliability (CR)']}")
    print(f"Cronbach's Alpha (CA): {result['Cronbach Alpha (CA)']}")
    print(f"Variance Inflation Factor (VIF):\n{result['Variance Inflation Factor (VIF)']}")
    print(f"Within Correlation: {result['Within Correlation']}\n")
```

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

def perform_convergent_analysis(data_, model_dict):
    """
    Perform convergent validity analysis for constructs in a given dataset.
    
    Parameters:
    - data_ (pd.DataFrame): Input dataset containing indicators.
    - model_dict (dict): Dictionary mapping constructs to their indicators.
    
    Returns:
    - pd.DataFrame: Unified DataFrame containing results (FL, AVE, CR, CA, VIF, Within Correlation) for all constructs.
    """
    # Output list to store results for all constructs
    output_data = []

    # Loop through each construct and its associated columns
    for construct, columns in model_dict.items():
        # Extract data for the construct
        construct_data = data_[columns]
        
        # Step 1: Standardize the data for VIF calculation
        scaler = StandardScaler()
        construct_data_scaled = scaler.fit_transform(construct_data)
        
        # Step 2: Compute Variance Inflation Factor (VIF) using improved method
        # Add small noise to avoid perfect multicollinearity
        construct_data_scaled += np.random.normal(0, 1e-5, construct_data_scaled.shape)
        vif = pd.DataFrame({
            "Variable": columns,
            "VIF": [variance_inflation_factor(construct_data_scaled, i)
                    for i in range(construct_data_scaled.shape[1])]
        })
        
        # Step 3: Conduct Exploratory Factor Analysis (EFA)
        fa = FactorAnalyzer(n_factors=1, rotation=None)  # 1-factor solution
        fa.fit(construct_data)
        
        # Extract factor loadings
        loadings = fa.loadings_
        factor_loadings = pd.DataFrame(loadings, index=columns, columns=["Factor1"])
        
        # Step 4: Compute Average Variance Extracted (AVE)
        squared_loadings = loadings**2
        AVE = np.mean(squared_loadings)
        
        # Step 5: Compute Composite Reliability (CR)
        loadings_sum = np.sum(loadings)
        CR = (loadings_sum**2) / ((loadings_sum**2) + np.sum(1 - squared_loadings))
        
        # Step 6: Compute Cronbach's Alpha (CA)
        k = construct_data.shape[1]  # Number of items
        item_variances = construct_data.var(axis=0, ddof=1)
        total_variance = construct_data.sum(axis=1).var(ddof=1)
        CA = (k / (k - 1)) * (1 - (item_variances.sum() / total_variance))
        
        # Step 7: Compute Within Correlation (HTMT Methodology)
        correlation_matrix = construct_data.corr()  # Compute pairwise correlations
        upper_triangle = correlation_matrix.where(
            np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
        )  # Extract upper triangle
        within_correlation = upper_triangle.stack().mean()  # Average of non-NaN values
        
        # Append results to output list
        for idx, row in factor_loadings.iterrows():
            output_data.append({
                'Construct': construct,
                'Item': idx,
                'Factor Loading': row['Factor1'],
                'AVE': AVE,
                'CR': CR,
                'CA': CA,
                'VIF': vif.loc[vif['Variable'] == idx, 'VIF'].values[0],
                'WiCorr': within_correlation
            })

    # Create a unified DataFrame for all constructs
    results_df = pd.DataFrame(output_data)
    return results_df

# Example Usage:
# Define model dictionary
demo_model_dict = {
    'GA': ['GA1', 'GA2', 'GA3', 'GA4', 'GA5'],
    'GL': ['GL1', 'GL2', 'GL3', 'GL4', 'GL5'],
    'GP': ['GP1', 'GP2', 'GP3', 'GP4'],
    'AT': ['AT1', 'AT2'],
    'PV': ['PV1', 'PV2', 'PV3'],
    'AW': ['AW1', 'AW2'],
    'PI': ['PI1', 'PI2', 'PI3']
}

# Perform convergent analysis
results_df = perform_convergent_analysis(data_, demo_model_dict)

# Display results
print("Convergent Validity Analysis Results:")
print(results_df["AVE"].sum())
results_df
```

```python
#regionCorrelation Matrix
import numpy as np
import pandas as pd


def calculate_correlation_matrix(data, constructs):
    """
    Calculate the correlation matrix among constructs.

    Parameters:
    - data (dict): Raw data as a dictionary with item names as keys and values as lists of data.
    - constructs (dict): A dictionary where keys are construct names and values are lists of item names.

    Returns:
    - pd.DataFrame: Correlation matrix among constructs.
    """
    # Convert data to DataFrame
    df = pd.DataFrame(data)

    # Extract construct labels
    construct_labels = list(constructs.keys())
    num_constructs = len(construct_labels)

    # Initialize correlation matrix
    correlation_matrix = np.zeros((num_constructs, num_constructs))

    # Calculate correlations
    for i, construct_i in enumerate(construct_labels):
        for j, construct_j in enumerate(construct_labels):
            if i == j:
                # Diagonal: Set to 1
                correlation_matrix[i, j] = 1.0
            else:
                # Get the items (indicators) for each construct
                indicators_i = constructs[construct_i]
                indicators_j = constructs[construct_j]

                # Compute pairwise correlations and average them
                correlations = [
                    df[indicator_i].corr(df[indicator_j])
                    for indicator_i in indicators_i
                    for indicator_j in indicators_j
                ]
                correlation_matrix[i, j] = np.mean(correlations)

    # Convert to DataFrame for better readability
    correlation_df = pd.DataFrame(correlation_matrix, columns=construct_labels, index=construct_labels)
    return correlation_df


# Calculate the correlation matrix
correlation_df = calculate_correlation_matrix(data_, model_dict)

print("Correlation Matrix:")
correlation_df

#endregion  
def calculate_fornell_larcker(correlation_df, results_df):
    """
    Calculate the Fornell-Larcker Criterion Matrix.

    Parameters:
    - correlation_df (pd.DataFrame): Correlation matrix between constructs.
    - results_df (pd.DataFrame): Results DataFrame containing AVE values for constructs.

    Returns:
    - pd.DataFrame: Fornell-Larcker Criterion Matrix.
    """
    # Extract AVE values for constructs
    ave = results_df.drop_duplicates(subset=['Construct'])[['Construct', 'AVE']].set_index('Construct')['AVE'].to_list()
    
    # Square root of AVE
    sqrt_ave = np.sqrt(ave)
    
    # Labels for constructs
    labels = correlation_df.columns

    # Create Fornell-Larcker Criterion matrix
    fornell_larcker = pd.DataFrame(correlation_df, columns=labels, index=labels)

    # Replace diagonal with sqrt(AVE)
    np.fill_diagonal(fornell_larcker.values, sqrt_ave)
    
    return fornell_larcker

# Example Usage:
# Define model dictionary
demo_model_dict = {
    'GA': ['GA1', 'GA2', 'GA3', 'GA4', 'GA5'],
    'GL': ['GL1', 'GL2', 'GL3', 'GL4', 'GL5'],
    'GP': ['GP1', 'GP2', 'GP3', 'GP4'],
    'AT': ['AT1', 'AT2'],
    'PV': ['PV1', 'PV2', 'PV3'],
    'AW': ['AW1', 'AW2'],
    'PI': ['PI1', 'PI2', 'PI3']
}

# Calculate Fornell-Larcker Criterion Matrix
fornell_larcker_matrix = calculate_fornell_larcker(correlation_df, results_df)

print("\nFornell-Larcker Criterion Matrix:")
fornell_larcker_matrix
```

```python
def calculate_htmt(correlation_df, results_df):
    """
    Calculate the HTMT (Heterotrait-Monotrait) Matrix.

    Parameters:
    - correlation_df (pd.DataFrame): Correlation matrix between constructs.
    - results_df (pd.DataFrame): Results DataFrame containing Within Correlations (WiCorr).

    Returns:
    - pd.DataFrame: HTMT Matrix.
    """
    # Extract construct order
    construct_order = results_df['Construct'].unique().tolist()

    # Reorder correlation matrix
    correlation_df_ordered = correlation_df.loc[construct_order, construct_order]
    correlation_matrix = correlation_df_ordered.to_numpy()

    # Extract Within-Correlations
    within_correlations = results_df.drop_duplicates(subset=['Construct'])[['Construct', 'WiCorr']].set_index('Construct')['WiCorr'].to_list()

    # Initialize HTMT matrix
    num_constructs = correlation_matrix.shape[0]
    htmt_matrix = np.zeros((num_constructs, num_constructs))

    # Calculate HTMT values
    for i in range(num_constructs):
        for j in range(num_constructs):
            if i != j:
                between_correlation = abs(correlation_matrix[i, j])  # Use absolute value for between-construct correlations
                htmt_matrix[i, j] = between_correlation / np.mean([abs(within_correlations[i]), abs(within_correlations[j])])  # Ensure within-correlations are absolute

    # Convert to DataFrame for readability
    htmt_df = pd.DataFrame(htmt_matrix, columns=construct_order, index=construct_order)
    return htmt_df
# Calculate HTMT Matrix
htmt_matrix = calculate_htmt(correlation_df, results_df)
htmt_matrix
```

```python
from semopy import Model
def fit_sem_model(data_, model_dict):
    """
    Fit a Structural Equation Model (SEM) using semopy.
    
    Parameters:
    - data_ (pd.DataFrame): Input dataset containing indicators.
    - model_dict (dict): Dictionary mapping constructs to their indicators.
    
    Returns:
    - pd.DataFrame: Model summary after fitting SEM.
    """
    # Dynamically build the measurement model portion
    measurement_model = ""
    for construct, indicators in model_dict.items():
        measurement_model += f"{construct} =~ {' + '.join(indicators)}\n"

    # Add the structural relationships
    structural_model = """
        AT ~ GA + GL + GP
        PV ~ GA + GL + GP
        AW ~ GA + GL + GP
        PI ~ AT + PV + AW + GA + GL + GP
    """

    # Combine the measurement and structural models
    model_desc = f"""
        {measurement_model}
        {structural_model}
    """

    # Fit the model using semopy
    model = Model(model_desc)
    model.fit(data_)

    # Summarize results
    report = model.inspect()
    return report

# Example Usage:
# Define model dictionary
demo_model_dict = {
    'GA': ['GA1', 'GA2', 'GA3', 'GA4', 'GA5'],
    'GL': ['GL1', 'GL2', 'GL3', 'GL4', 'GL5'],
    'GP': ['GP1', 'GP2', 'GP3', 'GP4'],
    'AT': ['AT1', 'AT2'],
    'PV': ['PV1', 'PV2', 'PV3'],
    'AW': ['AW1', 'AW2'],
    'PI': ['PI1', 'PI2', 'PI3']
}


# Perform SEM analysis
sem_report = fit_sem_model(data_, demo_model_dict)
print("\nSEM Model Summary:")


# Add a column to sem_report for significance levels based on p-value

# Add a column to sem_report for significance levels based on p-value
def add_significance_column(sem_report):
    """
    Add a significance level column to the sem_report DataFrame based on p-values.

    Parameters:
    - sem_report (pd.DataFrame): The semopy results DataFrame.

    Returns:
    - pd.DataFrame: The updated DataFrame with a significance level column.
    """
    def significance_level(p):
        try:
            p = float(p)
            if p <= 0.01:
                return 3
            elif p <= 0.05:
                return 2
            elif p <= 0.10:
                return 1
            else:
                return 0
        except ValueError:
            return 0

    sem_report['Significance Level'] = sem_report['p-value'].apply(significance_level)
    return sem_report


# Add significance level column to sem_report
sem_report = add_significance_column(sem_report)

sem_report = add_significance_column(sem_report)

sem_report
```

```python

def calculate_effects(df, predictor, mediator, outcome):

    direct_effect = df.loc[
        (df['lval'] == outcome) & (df['rval'] == predictor) & (df['op'] == '~'),
        'Estimate'
    ].values[0]

    # Extract the predictor-to-mediator effect (Predictor -> Mediator)
    pred_to_med = df.loc[
        (df['lval'] == mediator) & (df['rval'] == predictor) & (df['op'] == '~'),
        'Estimate'
    ].values[0]

    # Extract the mediator-to-outcome effect (Mediator -> Outcome)
    med_to_out = df.loc[
        (df['lval'] == outcome) & (df['rval'] == mediator) & (df['op'] == '~'),
        'Estimate'
    ].values[0]



    significance_1 = df.loc[
        (df['lval'] == outcome) & (df['rval'] == predictor) & (df['op'] == '~'),
        'Significance Level'
    ].values[0]
    
    significance_2 = df.loc[
        (df['lval'] == mediator) & (df['rval'] == predictor) & (df['op'] == '~'),
        'Significance Level'
    ].values[0]
    
    significance_3 = df.loc[
        (df['lval'] == outcome) & (df['rval'] == mediator) & (df['op'] == '~'),
        'Significance Level'
    ].values[0]
    
    # Calculate indirect effect (Predictor -> Mediator -> Outcome)
    indirect_effect = pred_to_med * med_to_out

    # Calculate total effect (Direct + Indirect)
    total_effect = direct_effect + indirect_effect

    # Create a DataFrame to store the results
    effects_df = pd.DataFrame({
        'Path': [f"{outcome} ~ {mediator} ~ {predictor}"],
        'Direct Effect': [direct_effect],
        'Indirect Effect': [indirect_effect],
        'Total Effect': [total_effect],
        'outcome_~_predictor': [significance_1],
        'outcome_~_mediator': [significance_3],
        'mediator_~_predictor': [significance_2]
    })

    return effects_df
def calculate_effect(df, predictors, mediators, outcome):
    """
    Test the mediating role of each mediator in the relationship between each predictor and the outcome.

    Parameters:
    - df (pd.DataFrame): DataFrame containing SEM results.
    - predictors (list): List of predictor variables.
    - mediators (list): List of mediator variables.
    - outcome (str): Outcome variable.

    Returns:
    - pd.DataFrame: Combined results of all mediation tests.
    """
    results = []
    for predictor in predictors:
        for mediator in mediators:
            effects_df = calculate_effects(df, predictor, mediator, outcome)
            results.append(effects_df)

    # Concatenate all results into a single DataFrame
    return pd.concat(results, ignore_index=True)

# Example usage
predictors = ['GA', 'GL', 'GP']
mediators = ['AT', 'AW', 'PV']
outcome = 'PI'


def check_mediation_role(effects_df):
    # Print columns to debug mismatches
    print("Columns in the DataFrame:", effects_df.columns)

    # Adjust column names if they don't match
    col_predictor_outcome = 'outcome_~_predictor'  # Update this to match your DataFrame
    col_predictor_mediator = 'mediator_~_predictor'  # Update this to match your DataFrame
    col_mediator_outcome = 'outcome_~_mediator'  # Update this to match your DataFrame

    def determine_mediation(row):
        predictor_to_outcome = row[col_predictor_outcome]
        predictor_to_mediator = row[col_predictor_mediator]
        mediator_to_outcome = row[col_mediator_outcome]

        # Apply mediation logic
        if predictor_to_mediator > 0 and mediator_to_outcome > 0:
            if predictor_to_outcome > 0:
                return 1  # Partial mediation
            else:
                return 2  # Full mediation
        else:
            return 0  # No mediation

    # Apply the logic to each row in the DataFrame
    effects_df['mediation'] = effects_df.apply(determine_mediation, axis=1)

    return effects_df

# Assuming `sem_report` is your DataFrame containing SEM results
effects_df_ = calculate_effect(sem_report, predictors, mediators, outcome)

effects_df = check_mediation_role(effects_df_)
effects_df
```
